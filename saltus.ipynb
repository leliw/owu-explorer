```python
import requests
from bs4 import BeautifulSoup
from requests.compat import urljoin, urlparse
```

```python
import json

def load_json_data(file: str):
    with open(file, 'tr', encoding="UTF-8") as json_file:
        data = json.load(json_file)
    return data

def save_json_data(file: str, json_data):
    with open(file, 'tw', encoding="UTF-8") as outfile:
        json.dump(json_data, outfile, indent=4, ensure_ascii=False)
```

```python
urls = set(['https://saltus.pl'])
skipped_domains = set(['itunes.apple.com', 'apps.apple.com', 'play.google.com', 'play.google.com', 'www.linkedin.com', 'www.youtube.com', 'www.amice-eu.org', 'www.gov.pl', 'piu.org.pl'])
max_depth = 3
```

```python
depth = 0
already_visited_url = set()
domains = set()
output = {}
```

```python
while depth < max_depth and len(urls) > 0:
    out_urls = set()
    for url in urls:
        try:
            response = requests.get(url)
            soup = BeautifulSoup(response.text, 'html.parser')
            already_visited_url.add(response.url)
            for link in soup.find_all('a'):
                url = urljoin(response.url, link.get('href'))
                domain = urlparse(url).netloc
                if not(domain in skipped_domains) and not(url in already_visited_url) and not(url in out_urls):
                    text = link.get_text().replace("\n", " ").strip()
                    if url.lower().endswith(".pdf") and text.lower().count("warunki") > 0:
                        print("{} {}: {} -> {}".format(depth, response.url, text, url))
                    out_urls.add(url)
                    domains.add(domain)
        except Exception as e:
            print(e)
    output[response.url] = { 'depth': depth, 'children': list(out_urls) }
    urls = out_urls
    depth += 1
print(sorted(domains))
save_json_data('saltus.json', output)
```
